```python
# Defining the function with multiple parameters
def myfunction(a,b,c):
    """This is just a function"""
    m=a*b**c
    n=b*c**a
    global number
    number=(m,n) #creating a tuple. this will also update the global copy of 'number'
    return number
```

```python
number=0
z=myfunction(2,3,4)
print(z)
a,b=z #unpacking the variables from the tuple. z[1] is also a valid call
print(a)
print(b)
print(number)
```

```python
## Very important concept- Nested functions that return function
def nestedfunction(a):
    """This function takes a number and returns another function that raises the number to the given power"""
    def raisepower(x):
        return x**a
    return raisepower
```

```python
square=nestedfunction(2)
cube=nestedfunction(3)
a=square(4)
b=cube(9)
print(a,b)
```

```python
#Function with flexible number of arguments
def flexible(*args):
    sum=0
    l=len(args)
    for num in args:
        sum+=num
        
    return (sum,l)
```

```python
print(flexible(1,2,3,4,5))
```

```python
# Creating function with Key-value arguments
def keyvaluefunction(**kargs):
    """This function takes any number key value pairs and performs some operations on it """
    for key,value in kargs.items():
        print(key,":",value)
```

```python
## defining ECDF function
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
## Plotting ECDF fuction
sb.set()
x,y=ecdf(iris.PetalLengthCm)
plt.scatter(x=x,y=y,c='red')
x,y=ecdf(iris.PetalWidthCm)
plt.scatter(x=x,y=y,c='Blue')
plt.show()
```

```python
#Summary statistics
np.mean(iris.PetalLengthCm)
np.percentile(iris.PetalLengthCm,[50,99])
```

```python
# Covariance and correlation
np.cov(iris.SepalLengthCm,iris.SepalWidthCm)
np.corrcoef(iris.SepalLengthCm,iris.SepalWidthCm)


## Random number generation
r=np.random.random(100)
x,y=ecdf(r)
plt.scatter(x,y,c='green')
##Binomial trials
b=np.random.binomial(6,0.5,size=100)
x,y=ecdf(b)
plt.scatter(x,y,c='blue')
## Poisson distribution
p=np.random.poisson(6,size=100)
x,y=ecdf(p)
plt.scatter(x,y,c='red')
plt.show()
```

```python
## Normal Distribution
n=np.random.normal(5,1,size=100000)
x,y=ecdf(n)

plt.scatter(x,y)
plt.show()
plt.hist(n,bins=10,histtype='step')
plt.show()
```
